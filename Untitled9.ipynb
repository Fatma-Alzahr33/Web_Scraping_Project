```python
import requests
from bs4 import BeautifulSoup
```

```python
link='https://market.isagha.com/prices'
page= requests.get(link)
print(page)
```

```python
soup=BeautifulSoup(page.text,('html.parser'))
print(soup)
rows=soup.find_all('tr')
GS_data=[]
for row in rows:
    pppp=row.find('span' , {'class':'purity-cell'})
    if pppp:
        gaa=pppp.text.strip()
        if "عيار" in gaa or'جنيه' in gaa : 
            all_tds= row .find_all('td')
            if len(all_tds)>=4:
                ss=all_tds[1].text.strip()
                gp=all_tds[3].text.strip()
                chaa=all_tds[-1].text.strip()
                td_html=str(row)
                if 'change_down' in td_html  or '-'in td_html  or ''in td_html  or '▼' in td_html :
                        status="هبوط"
                else:
                        status="صعود"
                        #print(gaa , ss ,gp,chaa ,status)
                blist=({ "العيار" :gaa ,
                          'سعر البيع': ss ,                                    
                          "سعر الشراء" : gp ,
                         "النسبه": chaa ,
                          "الحاله" : status })
                
                GS_data.append(blist)
                print(GS_data)                    
                

               
import pandas as pd 
df=pd.DataFrame(GS_data)
#print(df)

import csv
def printing():
    header=GS_data[0].keys()
    path='D:/golds_prices.csv'
    with open ( path , 'w' , encoding='utf-8-sig' , newline='') as file: 
        writer=csv.DictWriter(file, header)
        writer.writeheader()
        writer.writerows(GS_data)
        print('Done')
printing()
        
                

 
       
    
    
    
```
